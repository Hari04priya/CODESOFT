# TASK-3:IRIS FLOWER CLASSIFICATION

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
ds=pd.read_csv("IRIS[1].csv")
ds.head()
```

```python
x=ds[["sepal_length",	"sepal_width",	"petal_length",	"petal_width"]]
x.head()`
```

```python
y=ds["species"]
y.head()
```

```python
ds['species'],c=pd.factorize(ds['species'])
ds.head()
```

```python
ds.describe
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(ds.petal_length, ds.petal_width,ds.species)
ax.set_xlabel("petalLength (cm)")
ax.set_ylabel("petalWidth (cm)")
ax.set_zlabel("species")
plt.title("3D scattering plot for Petal_Length and Petal_Width ")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(ds.sepal_length, ds.sepal_width,ds.species)
ax.set_xlabel("sepalLength (cm)")
ax.set_ylabel("sepalWidth (cm)")
ax.set_zlabel("species")
plt.title("3D scattering plot for Sepal_Length and Sepal_Width")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_p=model.predict(x_test)
y_p
```

```python
conf_matrix=confusion_matrix(y_test,y_p)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
c_r=classification_report(y_test,y_p)
print("classifaction Report:")
print(c_r)
```

```python
accuracy=accuracy_score(y_test,y_p)
print('Accuracy:',accuracy*100)
```
